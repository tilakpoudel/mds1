```python
import pandas as pd
# Load the data
data = pd.read_csv('../../data/Diabetes.csv')
```

# Take data from diabetis data from sk learn

data.head()

```python
data.head()
data.info()
data.describe()
```

```python
# Check for missing values
data.isnull().sum()
```

```python
# split the data into features and target
x = data[['Glucose', "BMI"]]
y = data['Diabetes']
```

```python
# split the data into training and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
# view the test and train data
print(x_train.head())
print(x_test.head())
```

```python
# view the shape of the test and train data
print(x_train.shape)
print(x_test.shape)
```

```python
# view the test and train data for the target
print(y_train.head())
print(y_test.head())
```

```python
# now we need to fit the model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
# fit the model
model.fit(x_train, y_train)
```

```python
print(y_test)
```

```python
# make predictions on test data
y_pred = model.predict(x_test)
print(y_pred)
```

```python
# make prediction for a single data point
glucose = 183
bmi = 23.3
prediction = model.predict([[glucose, bmi]])
print(prediction)
```

```python
prediction_probabilty = model.predict_proba([[glucose, bmi]])
print(prediction_probabilty)
# the threshold is 0.5 by default, so model will predict 1 if the probability is greater than 0.5
# and 0 if the probability is less than 0.5
```

# Evaluate the model
## How do we evaluate logistic regression?
### 1. Accuracy
### 2. Precision
### 3. Recall
### 4. F1 score
### 5. Confusion matrix

We use confusion matrix to evaluate the logistic regression.

For linear model we use mean squared error, 


```python
## Use confusion matrix for logistic regression to evaluate the model.
from sklearn.metrics import confusion_matrix, accuracy_score
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
# Use classification report to evaluate the model
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
